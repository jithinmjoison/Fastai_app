```python
### Synthetic Data Generation ###

true_b = 1
true_w = 2
N = 100

np.random.seed(42)

x = np.random.rand(N,1)

epsilon = (0.1 * np.random.randn(N,1))

y = true_b + true_w * x + epsilon
```

```python
from re import I
### Shuffles the indexes###

idx = np.arange(N)

np.random.shuffle(idx)

### Train ###

train_idx = idx[:int(N*.8)]

val_idx = idx[int(N*.8):]

x_train = x[train_idx]

y_train = y[train_idx]

### Test ###

x_test = x[val_idx]

y_test = y[val_idx]
```

```python
### Gradient Descent ###

## Step 0 Random Intialisation ##


np.random.rand(42)
b = np.random.rand(1)
w = np.random.rand(1)

print(b,w)
```

```python
##Step 1 Forward Pass###

yhat = b+ w *  x_train

## Step 2 Loss

error = (yhat- y_train)

loss = np.mean(error **2)

### Computing the gradients for b and w since its a simple linear function

b_grad = 2 * error.mean()

w_grad = 2 * error.mean() * x_train

### Setting the learning rate

lr = 0.01

### Updating the parameters
b = b -lr * b_grad

w = w - lr * w_grad
```
